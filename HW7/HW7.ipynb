```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Define the nodes (concepts and methods)
nodes = [
    "Density Estimation",
    "Bayesian Decision",
    "Parametric Estimation",
    "Non-parametric Estimation",
    "Maximum Likelihood Estimation (MLE)",
    "Bayesian Estimation",
    "Histograms",
    "k-Nearest Neighbor Method",
    "Parzen-Window Method",
    "Bayesian Learning and Inference"
]

# Define the edges (relations)
edges = [
    ("Density Estimation", "Bayesian Decision"),
    ("Density Estimation", "Parametric Estimation"),
    ("Density Estimation", "Non-parametric Estimation"),
    ("Parametric Estimation", "Maximum Likelihood Estimation (MLE)"),
    ("Parametric Estimation", "Bayesian Estimation"),
    ("Non-parametric Estimation", "Histograms"),
    ("Non-parametric Estimation", "k-Nearest Neighbor Method"),
    ("Non-parametric Estimation", "Parzen-Window Method"),
    ("Bayesian Estimation", "Bayesian Learning and Inference"),
    ("Maximum Likelihood Estimation (MLE)", "Bayesian Learning and Inference")
]

# Add nodes and edges to the graph
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)  # for consistent layout
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=3000, edge_color='gray', linewidths=1, font_size=12)

# Display the graph
plt.title("Relations among Methods and Concepts in Density Estimation", size=15)
plt.show()
```
